```python
import pandas as pd
import imputation_toolkit as itk
import numpy as np
```

# Data Imputation:
In this jupyter notebook I show how to use the imputation_toolkit for two different datasets, namely the Iris data as a well known benchmark and the Sabina data, which was the actual purpose of this project.

## 1.Iris data:

### Loading the Iris data and dropping the class
Here we load a dataset for testing purpose and drop the categorical column.

```python
df = pd.read_csv("./iris-flower-dataset/IRIS.csv")
df = df.drop(["species"],axis=1)
df.head()
```

### Defining a dictionary of different imputation strategies and run the test:
We load the sklearn IterativeImputer and define a bunch of different imputation strategies. Afterwards we hand over the dictionary to the actual imputation testing function itk.test_imputation_methods() and define a miss_rate to check the quality of the imputation methods. For this step it's important that the testing df doesn't include any missing values to evaluate the result

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer, SimpleImputer       
from sklearn.linear_model import BayesianRidge, LinearRegression
from sklearn.tree import ExtraTreeRegressor



seed = 12

imputer_dict = {
    "mean": SimpleImputer(missing_values=np.nan, strategy="mean"),
    "median": SimpleImputer(missing_values=np.nan,strategy="median"),
    "BayesianRidge":IterativeImputer(estimator=BayesianRidge(),random_state=seed,initial_strategy = "mean"),
    "LinearRegression":IterativeImputer(estimator=LinearRegression(),random_state=seed,initial_strategy = "mean", max_iter=100),
    "ExtraTreeClassifier":IterativeImputer(estimator=ExtraTreeRegressor(random_state = seed),random_state=seed, initial_strategy="mean"),
    
}


df_results = itk.test_imputation_methods(df, imputer_dict, miss_rate = 0.3, seed = seed,save_Qplot=True)
df_results
```

The evaluation results of the imputation testing process shows that the best strategy is the iterative BayseanRidge regression with an R2 score of around 0.83.

## 2.Sabina data:

### Loading the Sabina data:

```python
import utils_imputer

df = utils_imputer.parse_folder_to_df(folder = "Sabina",datatype = "xls")
df.head()
```

### Preprocess the ONACE codes:
The ONACE code describes in which business sector a company is active in. The ONACE code is a hierarchical description, this means dropping a few of the last digits/letters of the code results in a more general sector description. In this case I drop all specifications except the first 3 digits/letters to reduce the overall amount of different ONACE codes. Whithout this preprocessing step we would encounter lots of instances with only one or a few samples.

```python
df["ONACE"] = df["ONACE 2008 Haupttätigkeit - Code"].apply(lambda x : str(x[:3]) if isinstance(x,str) else np.nan)
```

### Plotting the percentage of missing values per column:
Since the Sabina database has really a lot of different attributes, where the percentage of missing values is quite different, I will use the itk.bar_plot_missing_values function to get an overview.

```python
df_num_missing_values = itk.bar_plot_missing_values(df,save_fig=True)
```

From the plot we can obtain that the majority of the columns from the Sabina data have more than 50% missing values.

### Data preprocessing:
For this step I will first get ride of columns, where barely any values are not missing. Additionally I define a list of columns, which should be kept no matter how many values are missing.
To do so, I hand over the list to the itk.drop_cols_perc_na() function and drop all columns with more than 50% missing values.

```python
import imputation_toolkit as itk
cols = ['Umsatzerlöse\nEUR\nLetztes verf. Jahr','Umsatzerlöse\nEUR\n2018','Umsatzerlöse\nEUR\n2017','Umsatzerlöse\nEUR\n2016']
df_res = itk.drop_cols_perc_na(df, perc_missing=0.5, cols_to_keep = cols)
```

### Taking care of the categorical attributes:
Since the imputation_toolkit can only evaluate numerical attributes, we have to deal with non-numeric attributes beforehand. In case of the Sabina data the only non-numeric attribute is the already stemed ONACE code. To deal with this I use the sklearn.preprocessing.LabelEncoder(), with which I can easily transform the strings to numeric attributes.

```python
from sklearn import preprocessing
from tqdm import tqdm

df_res = df_res.dropna()
onace_encoder = preprocessing.LabelEncoder()
onace_encoder.fit(df_res["ONACE"].values)

df_res["ONACE"] = onace_encoder.transform(df_res["ONACE"])
df_res = df_res._get_numeric_data()
df_res.head()
```

### Test different imputation strategies for the Sabina data:
One more I define a dictionary of different imputation strategies and run the itk.test_imputation_methods() function. For this example I defined a miss_rate of 20%.

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer, SimpleImputer       
from sklearn.linear_model import BayesianRidge, LinearRegression
from sklearn.tree import ExtraTreeRegressor



seed = 12

imputer_dict = {
    "mean": SimpleImputer(missing_values=np.nan, strategy="mean"),
    "median": SimpleImputer(missing_values=np.nan,strategy="median"),
    "BayesianRidge":IterativeImputer(estimator=BayesianRidge(),random_state=seed,initial_strategy = "median"),
    "LinearRegression":IterativeImputer(estimator=LinearRegression(),random_state=seed,initial_strategy = "median", max_iter=100),
    "ExtraTreeClassifier_mean":IterativeImputer(estimator=ExtraTreeRegressor(random_state = seed),random_state=seed, initial_strategy="mean"),
    "ExtraTreeClassifier_median":IterativeImputer(estimator=ExtraTreeRegressor(random_state = seed),random_state=seed, initial_strategy="median"),
    
}

df_results = itk.test_imputation_methods(df_res, imputer_dict, miss_rate = 0.2, seed = seed,save_Qplot=True)
df_results
```

Also in case of the Sabina data the best performing imputation strategy is the iterative BayseanRidge regression, with an overall R2 score of 0.94. This is a pretty good result, especially if we compare it to standard methods like mean or median imputation, which result in a R2 score of around 1-3%.
